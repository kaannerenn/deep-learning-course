```python
#4) Aşağıdaki sözlükte, değerler içinde c harfinin geçip geçmediğini gösteren bir if koşulu yazınız
my_dictionary = {"k1":10,"k2k":"a","k32":30,"k4":"c"}
for values in my_dictionary.values():
    if type(values) == str:
        if "c" in values:
            print("c vardır")
        else:
            print("c yoktur.")
    else:
        print("c yoktur")
```

```python
#6) Aşağıdaki listedeki sayılardan sadece çift sayı olanları yazdıran bir kod yazınız.
my_numbers = [1,2,3,4,5,6,19,20,32,21,20,1111,23,24]
even_list = []
for i in my_numbers:
    if i % 2 == 0:
        even_list.append(i)
    else:
        continue
print(even_list)
```

```python
#7) Aşağıdaki listedeki sayılar bir dairenin yarı çapını vermektedir. 
#Tüm dairelerin çevresini içeren başka yeni bir liste oluşturunuz.
from math import pi
r_list = [3,2,5,8,4,6,9,12]
perimeter_list = []
for r in r_list:
    perimeter = r * pi * 2
    perimeter_list.append(perimeter)
print(perimeter_list)
```

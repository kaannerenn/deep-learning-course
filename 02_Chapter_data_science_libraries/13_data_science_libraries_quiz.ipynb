### Soru 2 
Aşağıdaki iki NumPy dizisi verilmiştir:

a = np.array([10, 20, 30, 40, 50])

b = np.array([1, 2, 3, 4, 5])

Bu iki dizi üzerinde eleman bazında toplama, çıkarma, çarpma ve bölme işlemlerini yapın. Örneğin 10 + 1 = 11 şeklinde index'i denk gelen tüm elemanlar birbiriyle işleme girip yeni dizide yerini almalıdır. Her sonucu ayrı satırda yazdırın.

```python
a = np.array([10, 20, 30, 40, 50])
b = np.array([1, 2, 3, 4, 5])
summatiaton = a + b
extraction = a - b
multiplication = a * b
divide = a / b
print(summatiaton,"---",extraction,"---",multiplication,"---",divide)
```

### Soru 7
Aşağıdaki iki DataFrame'i "id" sütununa göre birleştirin (merge) ve sonucu yazdırın:


Birleştirme sonrasında fiyatı 5000'den büyük olan ürünleri filtreleyin.

```python
df1 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "ürün": ["Laptop", "Telefon", "Tablet", "Kulaklık"]
})
df2 = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "fiyat": [15000, 10000, 5000, 500],
    "stok": [30, 50, 25, 100]
})
new_df = pd.merge(df1,df2,on="id",how="outer")
new_df
```

```python
new_df[new_df["fiyat"]>5000]
```

## Bölüm 3: Matplotlib

### Soru 8
Aşağıdaki verileri kullanarak bir çizgi grafiği (line plot) çizin:

Grafik gereksinimleri:

Başlık: "Aylık Satış Trendi"

X ekseni etiketi: "Aylar"

Y ekseni etiketi: "Satış Adedi"

Çizgi rengi kırmızı, işaretçi (marker) yuvarlak olsun

```python
aylar = ["Oca", "Şub", "Mar", "Nis", "May", "Haz"]
satislar = [120, 135, 148, 160, 175, 190]
plt.plot(aylar,satislar,"r",marker="o")
plt.xlabel("Aylar")
plt.ylabel("Satış Adedi")
plt.title("Aylık Satış Trendi")
plt.show()
```

### Soru 9
Aşağıdaki verileri kullanarak bir bar (çubuk) grafiği çizin:

Grafik gereksinimleri:

```python
diller = ["Python", "JavaScript", "Java", "C++", "Go"]
kullanici = [35, 25, 20, 12, 8]
plt.figure(figsize=(10,6))
plt.bar(diller,kullanici,color=["r","b","g","black","y"])
plt.title("Programlama dilleri popülerliği")
plt.show()
```

## Bölüm 4: Seaborn

### Soru 10

Seaborn'un yerleşik "tips" veri setini yükleyin

İlk 5 satırı gösterin.

total_bill sütunu için bir histogram (histplot) çizin.

Grafik başlığını "Toplam Hesap Dağılımı" olarak ayarlayın ve KDE eğrisini ekleyin.

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.set_style("whitegrid")
sns.histplot(data=df,y="total_bill",kde=True)
plt.title("Toplam Hesap Dağılımı")
plt.show()
```

### Soru 11
Tips veri setini kullanarak total_bill ve tip arasındaki ilişkiyi gösteren bir scatter plot (saçılım grafiği) çizin

Grafik gereksinimleri:

Noktaları smoker (sigara içen/içmeyen) kategorisine göre renklendirin (hue).

Başlık: "Hesap vs Bahşiş İlişkisi"

Figür boyutu 10×6 olsun

```python
df.columns
```

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.scatterplot(data=df,y="total_bill",x="tip",hue="smoker")
plt.title("Hesap ve Bahşiş İlişkisi")
plt.show()
```

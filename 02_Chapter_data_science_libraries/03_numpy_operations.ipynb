```python
# z-score hesaplama
data = np.array([10,12,13,15,18,25,100,105])
```

```python
mean = np.mean(data)
mean
```

```python
std = np.std(data)
std # standard sapma
```

```python
z_scores = (data - mean) / std
z_scores # outlier(aykırı) değerlerin skoru yüksek çıktı.
```

```python
z_scores[z_scores>1] # outlier değer gösterimi
```

```python
data[z_scores>1] # outlierlar
```

### mat equations

```python
# 2x + 3y = 8 and 5x + 7y = 19
```

```python
# coefficent matrix
coef = np.array([[2,3],[5,8]])
# constant matrix
cons = np.array([8,19])
# solution
solution = np.linalg.solve(coef,cons)
print(f"x:{solution[0]}-- y:{solution[1]}")
```

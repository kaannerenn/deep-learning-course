```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
"""Jupyter Notebook içinde matplotlib grafiklerinin doğrudan notebook hücresinin altında görünmesini
 sağlamak için yazılır."""
```

```python
age_list = [10,20,30,30,30,30,40,40,50,60,70,75]
weight_list = [20,60,70,70,80,90,95,92,74,55,67,68]
plt.plot(age_list,weight_list, "y") # y-> yellow
plt.title("Age vs Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.savefig("test.png") # Kaydedebiliriz.
plt.show()
```

#### Working with numpy

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
plt.plot(np_age_list,np_weight_list,"b")
plt.title("Age vs Weight")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```

```python
numpy_array1 = np.linspace(0,10,20) # 0 10 arası 20 tane sayı dön. Eşit aralıklarla
numpy_array1
```

```python
numpy_array2 = numpy_array1 ** 3
numpy_array2
```

```python
plt.plot(numpy_array1,numpy_array2,"g*-")
plt.show()
```

### Matplotlib eksenler

```python
plt.subplot(1,2,1) # 1 satır 2 sütunluk bir alan açalım ve 1.sine çiziyoruz anlamına gelir
plt.plot(numpy_array1,numpy_array2,"g*")
plt.subplot(1,2,2)
plt.plot(numpy_array2,numpy_array1,"b+-")
```

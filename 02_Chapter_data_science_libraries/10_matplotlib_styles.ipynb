```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
data2 = data1 ** 2
```

```python
my_fig, my_axes = plt.subplots()
print(type(my_fig),"***",type(my_axes))
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2,alpha=0.5) # Alpha -> transparanlığı sağlar.
my_axes.plot(data2,data1,"#1E0703") # Renk kodu da girebiliriz.
plt.show()
```

```python
(new_fig,new_axes) = plt.subplots()
new_axes.plot(data1, data1+2, color="blue",linewidth=1.0)
new_axes.plot(data1, data1+3, color="yellow",linewidth=3.0) # linewidth'te ayarlayabiliriz.
new_axes.plot(data1, data1+4, color="red",linestyle="-.") # linestyle depiştirebiliriz
new_axes.plot(data1, data1+5, color="green",linestyle=":")
new_axes.plot(data1, data1+6, color="black",linestyle="--")
new_axes.plot(data1, data1+7, color="pink",marker="o",markersize=8,markerfacecolor="black")
new_axes.plot(data1, data1+8, color="purple",marker="+",markersize=10)
plt.show()
```

```python
# scatter
plt.scatter(data1,data2)
plt.show()
```

```python
# histogram
new_arr = np.random.randint(0,100,50)
plt.hist(new_arr)
plt.show()
```

```python
# box plot
plt.boxplot(new_arr)
plt.show()
```

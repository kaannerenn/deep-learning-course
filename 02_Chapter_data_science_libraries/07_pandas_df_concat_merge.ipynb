### Merge

```python
df_merge1 = pd.read_csv("07-merge_data1.csv")
df_merge1
```

```python
df_merge2 = pd.read_csv("07-merge_data2.csv")
df_merge2
```

```python
# merge - outer join
df_merged_outer = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="outer") 
# on parametresi neye göre birleştiriceğimizi söyler.
# how parametreside hangi biçimde birleştiriceğimizi söyler.
# outer: 2 tabloyu birleştiriyor match edebildiğini yazdırıyor edemediklerine NaN veriyor.
df_merged_outer
```

```python
# merge - left join
df_merged_left = pd.merge(df_merge1,df_merge2,on="Employee_ID",how="left")
# Sadece 1. tabloyu alıp eğer onda 2. tabloyla eşleşen değerler varsa 1. tabloya o columnları ekliyor
# 2. tablodaki employee'leri gözardı etmiş oluyoruz
df_merged_left
```

```python
# merge - right join
df_merged_right = pd.merge(df_merge1,df_merge2, on="Employee_ID",how="right")
df_merged_right
# Buda leftteki mantıkla çalışıp 2. tabloya göre yapar aslında.
```

```python
# merge - inner join
df_merged_inner = pd.merge(df_merge1,df_merge2,on="Employee_ID",how="inner")
# Buda aslında kesişen değerleri alır. Hiç NaN değeri yoktur
df_merged_inner
```

## Series

```python
import numpy as np
import pandas as pd
```

```python
grades = {
    "Kaan": 50,
    "Eren": 70,
    "Yavuz": 35
}
pd.Series(grades)
```

```python
names = ["kaan","eren","yavuz"]
grades = [50,70,35]
print(pd.Series(names))
print()
print(pd.Series(grades))
print()
print(pd.Series(data=grades, index=names))
print(pd.Series(index=names, data=grades))
```

```python
# how to use with numpy
numpy_array = np.array([60,60,30,40])
pd.Series(numpy_array)
```

### series arithmetic


```python
contest_result = pd.Series(data=[10,20,30], index=["Kaan","Eren","Yavuz"])
contest_result2 = pd.Series(data=[50,60,80], index=["Kaan","Eren","Yavuz"])
print(contest_result["Kaan"],"---",contest_result2["Eren"])
```

```python
# Same indexes
final_result = contest_result + contest_result2
final_result2 = contest_result - contest_result2
final_result3 = contest_result / contest_result2
final_result4 = contest_result * contest_result2
print(final_result)
print(final_result2)
print(final_result3)
print(final_result4)
```

```python
# different indexes
different_series = pd.Series([20,30,40,50],["a","b","c","d"])
different_series2 = pd.Series([30,50,80,50],["a","c","f","g"])
print(different_series+different_series2)
# 2 sinde ortak olmayan datalar için NaN döner.
```

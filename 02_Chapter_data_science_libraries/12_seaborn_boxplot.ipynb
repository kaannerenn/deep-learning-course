```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Median, quartile(25%, 50%, 75%), min max 1.5 * IQR, outlier

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box plot")
plt.ylabel("Data Value")
plt.show()
# Ortadaki en uzun çizgi --> medyan
# Eşit uzunluktaki kısa üst ve alt çizgiler (whisker)--> en çok nereye çıktığı ve indiği değerlerin
# Yuvarlak olan 100'e denk gelen outlier değerdir.
```

```python
# seaborn'ün kendi hali hazırda yüklü kütüphaneleri vardır bu şekilde kullanabiliriz.
df = sns.load_dataset("titanic")
df
```

```python
plt.figure(figsize=(13,5))
plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.show()
```

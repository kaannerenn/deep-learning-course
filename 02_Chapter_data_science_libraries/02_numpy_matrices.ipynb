###  matrix arithmetic

```python
first_array = np.array([[10,20],[15,20]])
second_array = np.array([[5,15],[25,35]])
print(first_array+second_array) # Matrix summatiation
```

```python
third_array = np.array([[50,60]]) # 1 satır 2 sütunluk matris
print(first_array+third_array)
# Bu satırı first_array'in 2 satırıylada toplar. Normalde boyut uyumsuzluğu ama numpy bunu sağlıyor.
```

```python
fourth_array = np.array([[10,20,30,40,50]])
print(first_array+fourth_array)
# Bunu uyduramadığından typeError verir
```

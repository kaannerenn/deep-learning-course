### Decorators

```python
def my_decorator(func):

    
    def wrapper():
        print("wrapper function executed")
        func()
        print("wrapper function executed")
    
    return wrapper

@my_decorator
def hello_world():
    print("hello world")

hello_world()
```

### Property Decorators

```python
# data validation, private/public (encapsulation)

class Person:

    def __init__(self, name, age):
        self.__name = name # '__' ile private yaptık.
        self.__age = age

    # Getter
    @property
    def name(self):
        return self.__name
    
    # setter
    @name.setter
    def name(self,value):
        if value == "john":
            print("john not accepted")
        else:    
            self.__name = value

    
    # deleter
    @name.deleter
    def name(self):
        self.__name = None

kaan = Person(name="Kaan", age=20)
print(kaan.name)
kaan.name= "Kaan Eren Yavuz" # Bu işlem için @name.setter gereklidir.
print(kaan.name)
'''Setter ve getter gibi şeyleri kullanmamızın sebebi içine koşullar vs koyabilmemiz 
yoksa private yerine public yapar geçeriz, yani kuralları biz belirleriz.'''
kaan.name = "john"
print(kaan.name)
del kaan.name
print(kaan.name)
```

```python
class Person:

    def __init__(self, name, age):
        self.__name = name # '__' ile private yaptık.
        self.__age = age

    # Getter
    @property
    def name(self):
        return self.__name
    
    # setter
    @name.setter
    def name(self,value):
        
        if not isinstance(value,str):
            raise ValueError("Name must be a string.")
        if len(value) < 2:
            raise ValueError("name should be longer")
        self.__name = value

    
    # deleter
    @name.deleter
    def name(self):
        self.__name = None

    @property
    def age(self):
        return self.__age

    @age.setter
    def age(self,value):
        if not isinstance(value, int):
            raise ValueError("Age must be an integer")
        if value < 0:
            raise ValueError("Age should be a positive integer")
        self.__age = value
    

kaan = Person(name="Kaan", age=20)
# kaan.name = 70 # ValueError: Name must be a string.
#kaan.name = "a" # ValueError: name should be longer
kaan.name = "Kaan Eren Yavuz"
print(kaan.name)
#kaan.age = -1 # ValueError: Age should be a positive integer
#kaan.age = "asd" # ValueError: Age must be an integer
kaan.age = 30
print(kaan.age)
```

### Static Methods

### Class Methods

```python
# alternative constructor
class Pizza:

    total_pizzas = 0

    def __init__(self, ingredients):
        self.ingredients = ingredients
        Pizza.total_pizzas +=1

    @classmethod
    def margharita(cls):
        return cls(["peynir", "fesleğen","domates"])
    
    @classmethod
    def pepperoni(cls):
        return cls(["peynir", "sucuk","mısır"])
    
    @classmethod
    def get_total_pizzas(cls):
        return cls.total_pizzas
    
pizza1 = Pizza.margharita()
print(pizza1.ingredients)
pizza2 = Pizza.pepperoni()
print(pizza2.ingredients)
print(Pizza.get_total_pizzas())
```

### Abstract

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name


    @abstractmethod
    def make_sound(self):
        pass

    @abstractmethod
    def move(self):
        pass

    @abstractmethod
    def sleep(self):
        pass


class Dog(Animal):

    def make_sound(self):
        print("hav hav")

    def move(self):
        print("dog moved")

    def sleep(self):
        print("dog sleep")

barley = Dog("Barley")
barley.make_sound()
```

### overloading, overriding, final, typing

```python
def example(x: int):
    return x * 2

print(example(5))
print(example("kaan")) 
# bu da çalışır yani int verme zorunluluğumuz yok. Sadece okunurluğu artırırız.
```

```python
class Calculator:

    def add(self, a: int, b: int) -> int: # -> int geriye integer döndürücek demek.
        return a + b
    
    def add(self, a: int, b: int, c: int) -> int:
        return a + b + c
    
calculator = Calculator()
print(calculator.add(10,20,30)) # 2 method oluşturunca son oluşturulan geçerli.
# print(calculator.add(10,20)) hata verir
```

```python
from typing import overload, Union

class Calculator:

    @overload
    def add(self, a: int, b: int) -> int:
        ...
    
    @overload
    def add(self, a: int, b: int, c: int) -> int:
        ...

    def add (self,a: int, b: int, c: int | None = None) -> int:
        if c is None:
            return a + b
        return a + b + c
    
    @overload
    def process(self, value: int) -> int:
        ...

    @overload
    def process(self, value: str) -> str:
        ...

    def process(self, value: Union[int,str]) -> Union[int,str]:
        if isinstance(value, int):
            return value * 2
        elif isinstance(value, str):
            return value.capitalize()
    
calculator = Calculator()
print(calculator.add(10,20,30))
print(calculator.add(10,20))
print(calculator.process(10))
print(calculator.process("kaan"))
# Aslında overload'ta yine okunurluk artsın diye kullanmak zorunda değiliz.
```

### final, override

```python
from typing import final

class BaseGame:
    def start(self):
        print("game started")


    @final # override edilmeyecek fonksiyondur.
    def calculate_score(self, points: int) -> int:
        bonus = 100
        return points + bonus
    
    def end(self):
        print("game over")

class MyGame(BaseGame):

    # override ediyoruz start fonksiyonunu
    def start(self):
        print("MyGame started")

    # Aslında @final olan fonksiyonu override edebiliriz fakat bunu ayrıştırmak için @final kulanırız.
    # Yani @final kullandığımızı override etmek mantıklı bir yaklaşım değildir fakat çalışır.
    def calculate_score(self, points: int) -> int:
        return points * 3

# Sınıflarda @final kullanımıda bu sınıfı başka yerde kullanmak istemiyorum anlamında
# subcllass olarak vs kullanırsak uyarı verir hata vermez.
@final
class SecretAlgorithm:
    
    def process(self):
        print("secret algorithm used in game")


game = MyGame()
game.start()
game.calculate_score(100)
```

### override

```python
#from typing import override

class Shape:

    def area(self) -> float:
        return 0.0
    
    def perimeter(self) -> float:
        return 0.0
    
class Rectangle(Shape):

    def __init__(self, width: float, height: float):
        self.width = width
        self.height = height

    #@override
    def area(self) -> float:
        return self.width* self.height
    
    #@override
    def perimeter(self)-> float:
        return (self.width + self.height) * 2
    
rectangle = Rectangle(3,4)
print(rectangle.area())
print(rectangle.perimeter())

# ASlında override ettik ve override keywordlerinide kullanmadık python sürümünden ötürü import edemedik
# SAdece uyarı mesajı için yapıyoruz bunları aslında.
```

### Decorator Combining

```python
def multiply_decorator(func):

    def wrapper(x: int):
        return func(x) * 2
    return wrapper


def other_decorator(func):

    def wrapper(x: int):
        return func(x) * 4
    return wrapper


# Birden fazla decorator'u kullanabiliriz
@multiply_decorator
@other_decorator
def calculate(x: int):
    return x * 2

print(calculate(10))
```

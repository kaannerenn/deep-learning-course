### manipulating tensor: reshaping, stacking, squeezing, unsqueezing

```python
x = torch.arange(1,10,1)
x
```

```python
x.shape
```

```python
x.ndim
```

```python
# reshaping
x_reshaped = x.reshape(1,9)
x_reshaped
```

```python
x_reshaped.shape
```

```python
x_reshaped.ndim
```

```python
x_reshaped = x.reshape(3,3)
x_reshaped
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped
```

```python
x_reshaped[0] = 20
x_reshaped
```

```python
x # x_reshaped'teki ilk elemanı değiştiirnce x'teki ilk elemande değişiyor.
```

```python
# view
x_view = x.view(9,1)
x_view
```

```python
x_view[0] = 10
x_view
```

```python
x
```

```python
# view -> contigious tensorlarla çalışır sadece.
# non-contiguous: bellekte ardışık saklanmayan tensorlerdir.
# contiguous: bellekte ardışık (yan-yana) saklanan tensorlerdir.
```

```python
x = torch.tensor([[1,2,3],
                 [4,5,6]])
x.is_contiguous()
```

```python
y = x.t()
y
```

```python
y.is_contiguous()
```

### permute

```python
# sırayı değiştiriyoruz boyut kaybı yok.
x = torch.rand(size=(224,224,3))
x.shape
```

```python
x_permuted = x.permute(2,0,1) # axis 0->1 1->2 2->0
```

```python
x_permuted.shape
```

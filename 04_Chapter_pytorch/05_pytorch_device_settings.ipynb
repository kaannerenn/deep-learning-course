```python
import torch
```

```python
# requirements.txt dosyasından torch yüklendiğinde bazen default olarak cpu yüklenmesi gerçekleşebilir.
# böyle bir durumda torch ve torchvision kütüphanelerini silip manuel olarak terminalden gpu'lu versioynu yükleyebiliirz.
torch.cuda.is_available() # Bilgisayarda cuda ile çalışan bir gpu'muz var mı diye bakar.
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
device
```

### Tensorları taşıma

```python
tensor = torch.tensor([1,2,3])
```

```python
tensor
```

```python
tensor.device # default olarak cpu type'ında gelir.
```

```python
# manuel taşıma
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
tensor.numpy()
```

```python
tensor_on_gpu.numpy()
# tensorler gpu ise numpy'a çeviremeyiz. cpu olması lazım
```

```python
tensor_on_gpu.cpu().numpy() # cpu'ya cevirip numpy'a cevirebiliriz.
```

```python
# context manager kullanarak taşıma with keyboardı ile
with torch.device(device):
    tensor2 = torch.tensor([1,2,3])
    layer = torch.nn.Linear(20,30)
tensor2.device
# type olarak gpu'da oluştu
```

```python
layer.weight.device
# aynı şekilde buda gpu'da oluştu.
```

```python
tensor3= torch.tensor([4,5,6])
tensor3.device # with keyboard'ının dısında olusturdgumuz ıcın default olarak cpu ile oluşturuldu.
```

```python
# default'ını gpu olarak ayarladık fakat bundan önce oluşturulan cpu tensorlar gpu'ya dönmez
torch.set_default_device(device)
tensor4 = torch.tensor([1,2,3])
layer2 = torch.nn.Linear(10,20)
tensor4.device
```
